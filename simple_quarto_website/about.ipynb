```python
# INSERT YOUR SOLUTION HERE
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset from Seaborn
df = sns.load_dataset("penguins")
print(df.head())

# Creating initial plot
FirstPlot = sns.relplot(
data=df,
x='bill_length_mm',
y='flipper_length_mm',
hue='species',
col='island',
s=200
)

# Setting the theme of the graphs
sns.set_theme(style='white',palette="Pastel1")

# Printing the types
print('-----TYPES-----')
print(type(FirstPlot.fig))
print(type(FirstPlot.axes[0,0]))
print(type(FirstPlot.axes[0,1]))
print('                  ')

#setting the font size at 18
FS=18

# Axis Labels and Font Size
for i in range(0,3):
    FirstPlot.axes[0,i].set_xlabel('Penguin bill length (mm)', fontsize=FS)
FirstPlot.axes[0,0].set_ylabel('Penguin flipper length (mm)', fontsize=FS)


# Setting the titles
FirstPlot.axes[0,0].set_title("Torgersen Island", fontsize=FS)
FirstPlot.axes[0,1].set_title("Biscoe Island", fontsize=FS)
FirstPlot.axes[0,2].set_title("Dream Island", fontsize=FS)

#Getting the large ticks for the graphs
for i in range(0,3):
    FirstPlot.axes[0,i].tick_params(axis='both', which='major', labelsize=FS)

# Setting legend font size
plt.setp(FirstPlot._legend.get_texts(), fontsize=FS)
plt.setp(FirstPlot._legend.get_title(), fontsize=FS)

#Saving the figure
plt.savefig("plot-1.pdf", bbox_inches="tight")
plt.show()

```
